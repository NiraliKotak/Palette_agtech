```python
#List of independent and dependency countries: 'https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population'.
#List of UN recognized countries: 'https://en.wikipedia.org/wiki/List_of_sovereign_states'

import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population'
df_countries = pd.read_html(url)[1]
df_countries

un_countries = pd.read_html('https://en.wikipedia.org/wiki/List_of_sovereign_states')[1]
un_countries
```

```python
#Task 1:

#Read the first datasets and display the first 4 rows.
first_4_rows = df_countries.head(4)
print(first_4_rows)
#Convert the data types accordingly (e.g., float, string, integer) and perform data cleaning if necessary.

df_countries = df_countries.rename(columns={'Country / Dependency':'Country'})

df_countries['Country'] = df_countries['Country'].astype('string')
df_countries['Population'] = df_countries['Population'].astype('float')
df_countries.info()

#If the population is given in millions (e.g., Azerbaijan Population 10,000,000), divide it by 1,000,000 to represent it as 10.0.
df_countries['Population'] = df_countries['Population']/1000000
df_countries
```

```python
#Read the second dataframe and display the first 4 rows.
#Adjust the data types of the second dataframe as needed.

first_4_rows = un_countries.head(4)
print(first_4_rows)
un_countries = un_countries.rename(columns={'Common and formal names': 'Country'})
un_countries
un_countries['Country'] = un_countries['Country'].astype('string')
un_countries.info()
```

```python
#Merge dataframe 1 and dataframe 2 based on the country name. Ensure that the countries have the same name and can be merged together. You can use the following command: pd.merge(df2, df1, on='Country', how='inner')
un_countries
un_countries['Country'] = un_countries['Country'].str.replace(r'\s+–\s+[\w\s\']{1,}', '', regex=True)
un_countries
#print(df_countries)
#print(un_countries)

#merged_df = pd.merge(df1[['Country', 'Column1', 'Column2']], df2[['Country', 'Column3', 'Column4']], on='Country')

# Print the merged DataFrame

merge_data = pd.merge(df_countries[['Country', 'Population']], un_countries[['Country']], on='Country', how='inner')
merge_data
```
